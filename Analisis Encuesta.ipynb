```r
aciert_resp=NULL
for (i in 1:nrow(datos)){aciert_resp=rbind(aciert_resp,(cuestionario[i,]==resp_corr))}
head(aciert_resp)
```

```r
i=4
data.frame("clave"=resp_corr,"resp_docente"=t(cuestionario[i,]),"calificac"=aciert_resp[i,])
```

```r
aciertos_tipo=data.frame()#colnames=c("Sumativa","Formativa","Aprendizaje"))
for (i in 1:nrow(cuestionario)){
    for (resp in unique(resp_corr)){
    aciert_tipo=sum(aciert_resp[i,resp_corr==resp])/sum(resp_corr==resp)
    aciertos_tipo[i,resp]=aciert_tipo
    }
}
```

```r
round(colMeans(aciertos_tipo),3)
```
