```r
col_data=data[data$Country.Code=="COL",]
colnames(col_data)
```

```r
col_data$Indicator.Name
```

```r
col_data[col_data$Indicator.Name=="Mortality rate, under-5 (per 1,000)",]
```

```r
plot(y=as.numeric(col_data[col_data$Indicator.Name=="Mortality rate, under-5 (per 1,000)",5:ncol(col_data)]),x=c(1960:2016), type="l")
```

```r
mortalidad=data[data$Indicator.Code=="SH.DYN.MORT",]
```

```r
plot(y=rep(0,2016-1960+1),
     ylim=c(0,max(mortalidad[,c(5:ncol(mortalidad))],na.rm=t)),
     x=c(1960:2016),
     col="white",
     xlab="Año",
     ylab="Muertes por cada 1000 niños menores de 5 años"
)
for (i in 1:nrow(mortalidad)){
    indices=(mortalidad[i,c(5:ncol(mortalidad))])
    lines(x=c(1960:2016),y=indices, col=rgb(0,0,0,0.3))
}
lines(y=as.numeric(col_data[col_data$Indicator.Name=="Mortality rate, under-5 (per 1,000)",5:ncol(col_data)]),x=c(1960:2016), col="blue", lwd=2)
```
