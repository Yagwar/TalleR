# Análisis Escala de actitudes

```r
library(psych)
data=read.csv2("maye.csv")
test=data[,4:ncol(data)]
```

### Función alpha (para calcular el alfa de Cronbach)

https://www.personality-project.org/r/html/score.alpha.html

```r
alpha(test)
```

```r
for (i in 1:ncol(test)){
    print (paste("**************************","Item ",i,"**************************"))
    print (table(test[,i]))
    barplot(table(test[,i]), col="gray75", main=paste("Item ",i))
}
```
