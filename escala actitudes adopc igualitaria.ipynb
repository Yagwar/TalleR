```r
datos=read.csv2("adop_igual.csv", header = T)# cargar la base de datos original
cuest=datos[6:36]# separar de la base de datos las respuestas
```

```r
sentido=c("n","n","p","n","p","p","n","p","n","n","p","p","n","n","n","n","n","n","n","n","n","n","n","p","n","p","p","n","n","p","p")
```

```r
codif_neg=function(puntaje){
    return((puntaje-6)*-1)
}
```

```r
codif_neg(cuest$ITEM1)
```

```r
for (i in 1:ncol(cuest)){
    if (sentido[i]=="n"){
        cuest[i]=codif_neg(cuest[i])
    }
}
```

```r
Punt=data.frame(table(puntuac))$puntuac
fr_abs=data.frame(table(puntuac))$Freq
fr_abs_acum=cumsum(fr_abs)
fr_relat=data.frame(prop.table(table(puntuac)))$Freq
fr_relat_acum=cumsum(fr_relat)-((fr_relat)/2)
percentil=fr_relat_acum*100
punt_z=qnorm(fr_relat_acum)
punt_T=round(((punt_z)*15)+100,0)

baremo_2=data.frame(Punt,
                    fr_abs,
                    fr_abs_acum,
                    fr_relat,
                    fr_relat_acum,
                    percentil,
                    punt_z,
                    punt_T
                   )
baremo=data.frame(Punt,percentil,punt_z,punt_T)
baremo
```
